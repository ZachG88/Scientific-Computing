## Problem 2

```python
import numpy as np
import matplotlib.pyplot as plt
import math 

x = np.linspace(-np.pi,np.pi,100)
y = np.cos(x)
plt.plot(x,y,color = "black",linewidth = 2,label = "cos(x)") 

#Plot first order Taylor polynomial 
def taylor_first(x):
    temp = 0
    for l in range(2) :
       temp += (((-1)**l)/(math.factorial(2*l))) * (x**(2*l))  
    return temp 
z = taylor_first(x)  
plt.plot(x,z,color = "blue",linestyle = "dashed",linewidth = 2,label = "n = 1 Taylor")

#Plot third order Taylor polynomial 
def taylor_third(x) :
    temp = 0
    for l in range(4) :
       temp += (((-1)**l)/(math.factorial(2*l))) * (x**(2*l))  
    return temp 
p = taylor_third(x)
plt.plot(x,p,color = "red",linestyle = "dashdot",linewidth = 2, label = "n = 3 Taylor")

#Plot fourteenth order Taylor polynomial 
def taylor_fourteenth(x) :
    temp = 0
    for l in range(15) :
       temp += (((-1)**l)/(math.factorial(2*l))) * (x**(2*l))  
    return temp 
q = taylor_fourteenth(x)
plt.plot(x,q,color = "magenta",linestyle = "dotted",linewidth = 2,label = "n = 14 Taylor")

plt.title("cos(x) and its Taylor approximations")
plt.xlabel("x-values")
plt.ylabel("cos(x) approximations")
plt.legend(loc="center")
plt.show()
```
